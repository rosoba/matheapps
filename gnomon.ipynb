## Gnomonfigur
Die Bezeichung der Seiten: $g$ = 10, $h$ = 6. Die Fläche $F$ und Umfang $U$ des Rechecks:
$ F = gh$, \; $ U = 2(g+h)$

```python
%matplotlib notebook
#import sympy as sp # symbolic algebra package
import numpy as np # numerical package
import matplotlib.pyplot as plt # plotting package
#sp.init_printing() # enable nice formating of the derived expressions
import ipywidgets as ipw
```

```python
h, g = 6., 10.
f_max = g*0.2

fig, ax = plt.subplots(1,1, figsize=(9,4))
ax.set_xlim(-f_max, g+f_max )
base, = ax.fill([0, g, g, 0], [0, 0, h, h], color='yellow', alpha=1)
ax.axis('equal')
eta = 0.4
diag, = ax.plot([0,g],[0,h], color='red', lw=3)
line_h, = ax.plot([0,0],[0,h], color='blue', lw=4)
line_g, = ax.plot([0,g],[0,0], color='green', lw=4)
bl, = ax.fill([0,eta*g, eta*g, 0],[0, 0, eta*h, eta*h], color='gray', alpha=0.2)
tr, = ax.fill([eta*g, g, g, eta*g],[eta*h, eta*h, h, h], color='gray', alpha=0.2)
def update(eta, f):
    gf = g+f
    vertexes = np.array([[0,g, gf, f], [0, 0, h, h]], dtype=np.float_)
    base.set_xy(vertexes.T)
    vertexes = np.array([[0,eta*g, eta*(gf), eta*f], [0, 0, eta*h, eta*h]], dtype=np.float_)
    bl.set_xy(vertexes.T)
    vertexes = np.array([[eta*gf, g+eta*f, gf, eta*g+f],[eta*h, eta*h, h, h]], dtype=np.float_)
    tr.set_xy(vertexes.T)
    diag.set_data([0, gf],[0, h])
    line_h.set_data([eta*g, eta*g+f],[0, h])
    line_g.set_data([eta*f, g+eta*f],[eta*h, eta*h])
    print('F1: %g * %g = %g' % (eta*g, (1-eta)*h, eta*g * (1-eta)*h))
    print('U1: 2*(%g + %g) = %g' % (eta*g, (1-eta)*h, 2*(eta*g + (1-eta)*h)))
    print('F2: %g * %g = %g' % ((1-eta)*g, eta*h, (1-eta)*g * eta*h))
    print('U2: 2*(%g + %g) = %g' % ((1-eta)*g, eta*h, 2*((1-eta)*g + eta*h)))

ipw.interact(update, 
            eta=ipw.FloatSlider(description='Schieber', 
                                min=0, max=1, step=0.1),
            f=ipw.FloatSlider(description='Neigung', 
                              min=-f_max, max=f_max, step=f_max/10)
            );
```
